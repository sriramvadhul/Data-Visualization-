```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style="whitegrid")

file_path = r"C:\Users\Dell\Desktop\Masters\2nd Sem\Data Visulization\Final Project\Final Project\medical_data.csv"
df = pd.read_csv(file_path)

df.head()
```

```python
df.info()
df.describe()
```

```python
plt.figure(figsize=(10,4))
sns.countplot(
    data=df,
    x="diagnosis",
    order=df["diagnosis"].value_counts().index
)
plt.xticks(rotation=45)
plt.title("Diagnosis Distribution in Patient Population")
plt.show()
```

```python
# - Insight

#The dataset reflects realistic clinical imbalance, with non-critical diagnoses occurring more frequently than acute conditions.
```

```python
plt.figure(figsize=(10,4))
sns.boxplot(data=df, x="diagnosis", y="age")
plt.xticks(rotation=45)
plt.title("Age Distribution by Diagnosis")
plt.show()
```

```python
#Severe conditions such as myocardial infarction and sepsis are concentrated among older patients.
```

```python
vitals = ["temperature", "heart_rate", "respiratory_rate", "oxygen_saturation"]

for v in vitals:
    plt.figure(figsize=(9,4))
    sns.boxplot(data=df, x="diagnosis", y=v)
    plt.xticks(rotation=45)
    plt.title(f"{v.replace('_',' ').title()} by Diagnosis")
    plt.show()
```

```python
#Oxygen saturation and respiratory rate provide stronger diagnostic separation than temperature alone.
```

```python
labs = ["wbc", "crp"]

for lab in labs:
    plt.figure(figsize=(9,4))
    sns.boxplot(data=df, x="diagnosis", y=lab)
    plt.xticks(rotation=45)
    plt.title(f"{lab.upper()} Levels by Diagnosis")
    plt.show()
```

```python
#Elevated inflammatory markers are strongly associated with systemic infections.
```

```python
plt.figure(figsize=(6,5))
sns.scatterplot(
    data=df.sample(2000),
    x="wbc",
    y="crp",
    hue="diagnosis",
    alpha=0.6,
    legend=False
)
plt.title("Relationship Between WBC and CRP")
plt.show()
```

```python
#Joint elevation of WBC and CRP is indicative of more severe inflammatory conditions.
```

```python
df["comorbidity_count"] = df[
    ["diabetes", "hypertension", "asthma", "smoker"]
].sum(axis=1)

plt.figure(figsize=(9,4))
sns.barplot(
    data=df,
    x="diagnosis",
    y="comorbidity_count",
    estimator=np.mean
)
plt.xticks(rotation=45)
plt.title("Average Comorbidity Burden by Diagnosis")
plt.show()
```

```python
#Chronic disease burden increases noticeably in severe diagnoses, reinforcing cumulative risk effects.
```

```python
respiratory = [
    "COVID-19","Pneumonia","Bronchitis",
    "Asthma Exacerbation","Influenza"
]

df["disease_group"] = df["diagnosis"].apply(
    lambda x: "Respiratory" if x in respiratory else "Non-Respiratory"
)

plt.figure(figsize=(6,4))
sns.boxplot(data=df, x="disease_group", y="oxygen_saturation")
plt.title("Oxygen Saturation by Disease Group")
plt.show()
```

```python
#Respiratory diagnoses consistently show reduced oxygen saturation, validating physiological grouping.
```

```python
### Conclusion
This analysis demonstrates that integrating vitals and laboratory data significantly enhances diagnostic insight compared to symptom-only analysis. Oxygen saturation, inflammatory markers, and comorbidity burden emerge as key indicators of disease severity.
```

```python
# The population spans all age groups, allowing age-stratified analysis of disease risk.
```

```python
#Cardiovascular and systemic conditions tend to affect older patients, while neurological conditions skew younger.
```
